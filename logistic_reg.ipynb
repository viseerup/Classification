A Logistic regression classifier consist of 3 distinct parts 

1. A standard (possibly) multivariate linear function :  $ z = \mathbf{\theta}^\top\mathbf{x} = \sum_i W_i*x_i +b $

2. An activation function, most commonly the sigmoid function $\sigma(z)=\frac{1}{1+\text{e}^{(-z)}}$ 

3. A binary step function at $\sigma(z)>0.5$ .
